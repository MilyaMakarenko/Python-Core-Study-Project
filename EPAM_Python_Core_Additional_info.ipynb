```python
class Person:
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

    def __str__(self):
        return f'This is a person with name {self.name}  and he/she is {self.age} years old!'
    
    # встроенный метод сложения
    def __add__(self, another_person):
        return Person(self.name + another_person.name , self.age + another_person.age)
    # встроенный метод сравнение на равность
    def __eq__(self, another_person):
        return self.name == another_person.name and self.age == another_person.age
    # встроенный метод возведения в степень
    def __pow__(self, indicator):
        return Person(f'{self.name}^{indicator}', self.age ** indicator)
    # свой встроенный метод, и они не вызыываются по умолчанию, придется вызывать явно
    def __pylounge__(self):
        return f'{self.name}: {self.age}'
```

```python
maxik = Person('Maks', 5)
evan = Person('Evan', 24)

print(maxik + evan)
```

```python
maxik_clone  = Person('Maks', 5)
print(maxik == maxik_clone)
```

```python
print(maxik**3)
print(maxik.__pylounge__())
```

```python
def get_my_links_decorator(func):
    def wrapper():
        print ('My current links')
        func()
        print("Don't hesitate to text me!")
    return wrapper
```

```python
@get_my_links_decorator
def get_my_links():
    print( {'LinkedIn': 'https://www.somelink',
            'GitHub': 'https://github.com/somelink',
            'Stepik': 'https://stepik.org/users/somelink'
            }
    )
```

```python
def get_my_links_decorator(func):
    def wrapper(*args, **kwargs):
        print ('My current links')
        func(*args, **kwargs)
        print("Don't hesitate to text me!")
    return wrapper
```

```python
# декаратор с аргументами
def outer_decorator(hy_msg = 'My current links', bye_msg = "Don't hesitate to text me!"):
    def inner_decorator(func):
        def wrapper(*args, **kwargs):
            print (hy_msg)
            func(*args, **kwargs)
            print(bye_msg)
        return wrapper
    return inner_decorator
```

```python
@outer_decorator('Hi', 'Bye!!!!')
def get_my_links():
    print( {'LinkedIn': 'https://www.somelink',
            'GitHub': 'https://github.com/somelink',
            'Stepik': 'https://stepik.org/users/somelink'
            }
    )


get_my_links()
```

```python
def i_am_top(func):
    def wrapper(*args, **kwargs):
        print('I am on top!')
        func(*args, **kwargs)
    return wrapper
```

```python
class Add:
    x = 9

    def __init__(self, x):
        self.x = x
    
    @property
    def sqr_x(self):
        return self.x ** 2 
    
    def addMethod(self, y):
        print("method:", self.x + y)

    @classmethod
    def addClass(self, y):
        print('claddmethod:', self.x + y)

    @staticmethod
    def addStatic(self, y):
        print(f"staticmethod: {self.x + y}")

a =Add(2)
print(a.sqr_x)
```

```python
print(a.addStatic(a, 4))
```

```python
from typing import List


def sum_numbers(num1: float = 0, num2: int = 0):
    return num1 + num2

def sum_strings(str1: str = '', str2: str = ''):
    return f'{str1}:{str2}'

def sum_list(lst1: List, lst2: List):
    return [*lst1, *lst2]

arg1 = [1, 2, 3]
arg2 = [4, 5, 6]

if isinstance(arg1, list):
    print(sum_list(arg1, arg2))
elif isinstance(arg1, int):
    print(sum_numbers(arg1, arg2))
else:
    print(sum_list(arg1, arg2))
```

```python
class Summarizer:
    def __init__(self):
        self.result = None
    def sum(self, arg1, arg2):
        if isinstance(arg1, list):
            self.result = [*arg1, *arg2]
        elif isinstance(arg1, int):
            self.result = arg1 + arg2
        else:
            self.result = f'{arg1}:{arg2}'
        return self.result 
    
arg1 = [1, 2, 3]
arg2 = [4, 5, 6]
sum_master = Summarizer()
print(sum_master.sum(arg1, arg2))
```
