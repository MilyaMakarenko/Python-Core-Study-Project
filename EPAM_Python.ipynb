```python
def get_fib(n):
    n_fib = None
    
    if n < 1: # check that input is a valid
        print('N must be > 0')
    elif n == 1: # first Fib number is 0
        n_fib = 0
    elif n == 2: # second number - 1
        n_fib = 1
    else:
        prev_2, prev_1 = 0, 1        # prev_2 – N-2 element, prev_1 – N-1
        for i in range(2, n):
            n_fib = prev_2 + prev_1  # calculate a next value of the Fib
            prev_2 = prev_1          # shift prev_2 and prev_1 values
            prev_1 = n_fib
    return n_fib
```

```python
from typing import Dict
from time import sleep
import time

execution_time: Dict[str, float] = {}

def time_decorator(fn):
    """
    Create a decorator function `time_decorator`
    which has to calculate decorated function execution time
    and put this time value to `execution_time` dictionary where `key` is
    decorated function name and `value` is this function execution time.
    """
    def wrapper(*args, **kwargs):
        # Засекаем время до выполнения функции
        start_time = time.time()
        
        # Выполняем функцию и сохраняем результат
        result = fn(*args, **kwargs)
        
        # Засекаем время после выполнения функции
        end_time = time.time()
        
        # Вычисляем время выполнения
        execution_time[fn.__name__] = end_time - start_time
        
        # Возвращаем результат функции
        return result
    
    return wrapper

@time_decorator
def func_add(a, b):
    sleep(0.2)
    return a + b

print(func_add(10, 20))


print(execution_time['func_add'])
```

```python
class Counter:
    def __init__(self, start=0, stop=None):
        self.current = start
        self.stop = stop
    
    def increment(self):
        # Проверяем, не достигнут ли лимит
        if self.stop is not None and self.current >= self.stop:
            print("Maximal value is reached.")
        else:
            self.current += 1
    
    def get(self):
        return self.current
```

```python
class HistoryDict:

    def __init__(self, initial_dict={}):
        self.dicts = initial_dict
        self.res_list = []

    def set_value(self, key, value):
        self.dicts[key] = value
        self.res_list.append(key)
        if len(self.res_list) > 5:
            self.res_list = self.res_list[:-5]

    def get_value(self, key):
        return self.dicts[key]


    def get_history(self):
        return self.res_list
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC): 
 
    def __init__(self, brand_name: str, year_of_issue: int, base_price: int, mileage:int ): 
        self.brand_name = brand_name
        self.year_of_issue = year_of_issue
        self.base_price = base_price
        self.mileage = mileage

    @abstractmethod
    def vehicle_type(self) -> str:
        """Должен возвращать brand_name + название класса"""
        return f"{self.brand_name} {self.__class__.__name__}"
    

    @abstractmethod
    def is_motorcycle(self) -> bool:
        return (self == 2)
 
    @property
    @abstractmethod
    def purchase_price(self) -> float:
        price = self.base_price - 0.1 * self.mileage
        return max(price, 100_000)

class Car(Vehicle):
    def vehicle_type(self) -> str:
        return f"{self.brand_name} Car"
    
    def is_motorcycle(self) -> bool:
        return False
    
    @property
    def purchase_price(self) -> float:
        price = self.base_price - 0.1 * self.mileage
        return max(price, 100_000)

class Motorcycle(Vehicle):
    def vehicle_type(self) -> str:
        return f"{self.brand_name} Motorcycle"
    
    def is_motorcycle(self) -> bool:
        return True  # У мотоцикла 2 колеса
    
    @property
    def purchase_price(self) -> float:
        price = self.base_price - 0.1 * self.mileage
        return max(price, 100_000)

class Truck(Vehicle):
    def vehicle_type(self) -> str:
        return f"{self.brand_name} Truck"
    
    def is_motorcycle(self) -> bool:
        return False
    
    @property
    def purchase_price(self) -> float:
        price = self.base_price - 0.1 * self.mileage
        return max(price, 100_000)

class Bus(Vehicle):
    def vehicle_type(self) -> str:
        return f"{self.brand_name} Bus"
    
    def is_motorcycle(self) -> bool:
        return False
    
    @property
    def purchase_price(self) -> float:
        price = self.base_price - 0.1 * self.mileage
        return max(price, 100_000)
```

```python
class Pagination:
    def __init__(self, data: str, items_on_page: int):
        self.data = data
        self.items_on_page = items_on_page
        self.pages = []

        # Разбиваем текст на страницы при инициализации
        for i in range(0, len(self.data), self.items_on_page):
            self.pages.append(self.data[i:i + self.items_on_page])

    @property
    def page_count(self):
        """Возвращаем количество страниц"""
        return len(self.pages)
        # cnt = len(self.data) // self.items_on_page
        # cnt_list = []
        # for i in range(0, len(self.data), self.items_on_page):
        #     cnt_list.append(self.data[i:i+self.items_on_page])

    @property
    def item_count(self):
        """Возвращаем общее количество символов в тексте"""
        return len(self.data)



    def count_items_on_page(self, page_number):
        """Возвращаем количество символов на указанной странице"""
        if page_number < 0 or page_number >= self.page_count:
            raise Exception("Invalid index. Page is missing.")
        return len(self.pages[page_number])

    def find_page(self, search_term: str):
        """Ищет страницы, содержащие указанный символ или слово"""
        result = []
    
        # Проверяем каждую страницу
        for i, page in enumerate(self.pages):
            # Проверяем, есть ли часть искомого слова на этой странице
            if any(part in page for part in [search_term, 
                                            search_term[:len(page)], 
                                            search_term[-len(page):]]):
                result.append(i)
            
            # Проверяем склейку текущей страницы со следующей
            if i < len(self.pages) - 1:
                combined = page + self.pages[i + 1]
                if search_term in combined:
                    if i not in result:
                        result.append(i)
                    if i + 1 not in result:
                        result.append(i + 1)
        
        result = sorted(list(set(result)))
        
        if not result:
            raise Exception(f"'{search_term}' is missing on the pages")
        
        return result
    
    def display_page(self, page_number):
        """Отображает содержимое указанной страницы"""
        if page_number < 0 or page_number >= self.page_count:
            raise Exception("Invalid index. Page is missing.")
        return self.pages[page_number]
    
```

```python
from typing import Union


def divide(str_with_ints: str) -> Union[float, str]:
    """
    Returns the result of dividing two numbers or an error message.
    :arg
        str_with_ints: str, ex. "4 2";
    :return
        result of dividing: float, ex. 2.0 (4 / 2 = 2.0);
        error response in "Error code: {error message}: str;
    """
    try:
        a = int(str_with_ints.split(' ')[0])
        b = int(str_with_ints.split(' ')[-1])
        res = None



        try:
            res = a / b
        except ZeroDivisionError:
            return "Error code: division by zero"
            
    except ValueError:
        return "Error code: invalid literal for int()"

    return res
    


# >>> divide("4 2")
# 2.0

# >>> divide("4 0")
# "Error code: division by zero"

# >>> divide("* 1")
# "Error code: invalid literal for int() with base 10: '*'"
```

```python
class FiniteRepeat:
   def __init__(self, msg, max_count):
     self.msg = msg
     self.max_count = max_count
     self.count = 0
     
   def __iter__(self):
     return self
     
   def __next__(self):
     if self.count >= self.max_count:
        raise StopIteration
     self.count += 1
     return self.msg

obj = FiniteRepeat("car", 5)
for message in obj:
   print(message)
```

```python
class Bird:
    def __init__(self, name):
        self.name = name

    def fly(self):
        return f"{self.name} bird can fly"


    def walk(self):
        return f"{self.name} bird can walk"
    
    def __str__(self):
        return f"{self.name} is a bird"


class FlyingBird(Bird):
    def __init__(self, name, ration="grains"):
        super().__init__(name)
        self.ration = ration


    def eat(self):
        return f"It eats mostly {self.ration}"

    def __str__(self):
        return f"{self.name} bird can walk and fly"




class NonFlyingBird(Bird):
    def __init__(self, name, ration="fish"):
        super().__init__(name)
        self.ration = ration

    def fly(self):
        return f"AttributeError: '{self.name}' object has no attribute 'fly'"
    
    def walk(self):
        return f"{self.name} bird can walk"
    
    def swim(self):
        return f"{self.name} bird can swim"
    
    def eat(self):
        return f"It eats mostly {self.ration}"
    
    def __str__(self):
        return f"{self.name} is a non-flying bird"
    
class SuperBird(NonFlyingBird, FlyingBird):
    def __init__(self, name, ration="fish"):
        # Используем __init__ от NonFlyingBird для инициализации
        NonFlyingBird.__init__(self, name, ration)
    
    def fly(self):
        return f"{self.name} bird can fly"
    
    def swim(self):
        return f"{self.name} bird can swim"
    
    # Метод eat наследуется от NonFlyingBird (первый в порядке наследования)
    # Благодаря MRO (Method Resolution Order)
    
    def __str__(self):
        return f"{self.name} bird can walk, swim and fly"
```

```python
class Currency:
    def __init__(self, value):
        self.value = value

    @classmethod   
    def course(cls, other_cls):
        """Возвращает курс для конвертации"""
        # Получаем курсы относительно EUR
        rates = {
            'Euro': 1.0,
            'Dollar': 2.0,   # 1 EUR = 2 USD
            'Pound': 100.0    # 1 EUR = 100 GBP
        }
        # Конвертируем через EUR
        if cls.__name__ == 'Euro':
            rate = rates[other_cls.__name__]
        elif other_cls.__name__ == 'Euro':
            rate = 1.0 / rates[cls.__name__]
        else:
            # Конвертация через EUR: сначала в EUR, потом в целевую валюту
            rate = rates[other_cls.__name__] / rates[cls.__name__]
        
        return f"{rate} {other_cls.get_symbol()} for 1 {cls.get_symbol()}"

    @staticmethod
    def get_symbol():
        """Возвращает символ валюты (должен быть переопределен в наследниках)"""
        return ""
    
    def to_currency(self, other_cls):
        """Конвертирует текущую валюту в другую"""
        # Получаем курсы относительно EUR
        rates = {
            'Euro': 1.0,
            'Dollar': 2.0,
            'Pound': 100.0
        }
        
        # Конвертируем через EUR
        value_in_eur = self.value / rates[self.__class__.__name__]
        new_value = value_in_eur * rates[other_cls.__name__]
        
        return other_cls(new_value)
    
    def __add__(self, other):
        """Сложение валют"""
        # Конвертируем другую валюту в текущую и складываем
        converted = other.to_currency(self.__class__)
        return self.__class__(self.value + converted.value)
    
    def __gt__(self, other):
        """Сравнение >"""
        # Конвертируем другую валюту в текущую для сравнения
        converted = other.to_currency(self.__class__)
        return self.value > converted.value
    
    def __lt__(self, other):
        """Сравнение <"""
        converted = other.to_currency(self.__class__)
        return self.value < converted.value
    
    def __eq__(self, other):
        """Сравнение =="""
        converted = other.to_currency(self.__class__)
        return abs(self.value - converted.value) < 1e-10  # Учитываем погрешность float
    
    def __str__(self):
        return f"{self.value} {self.get_symbol()}"
    


class Euro(Currency):
    @staticmethod
    def get_symbol():
        return "EUR"

class Dollar(Currency):
    @staticmethod
    def get_symbol():
        return "USD"
    
class Pound(Currency):
    @staticmethod
    def get_symbol():
        return "GBP"
```
